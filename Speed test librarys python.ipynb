```python
import time
start_time = time.time()

from textblob import TextBlob

texto = "NLP, o Procesamiento del Lenguaje Natural, es una subdisciplina de la inteligencia artificial que se enfoca en la interacción entre las computadoras y el lenguaje humano. Se trata del desarrollo de algoritmos y técnicas que permiten a las computadoras entender, interpretar y generar lenguaje natural, así como interactuar con los seres humanos de una manera natural."
blob = TextBlob(texto)

# Tokenización de palabras
palabras = blob.words
cont=0
print("Palabras:")
for palabra in palabras:
    print(palabra)
    cont=cont+1
end_time = time.time()
print("\n Tiempo:", end_time - start_time, "seconds")
esperado=63
efect=((cont/esperado)*100)
print("palabras encontradas:",cont)
print("palabras esperadas:", esperado)
print("Efectivdad de tokenización", efect)
```

```python
import time
start_time = time.time()
import spacy

texto = "NLP, o Procesamiento del Lenguaje Natural, es una subdisciplina de la inteligencia artificial que se enfoca en la interacción entre las computadoras y el lenguaje humano. Se trata del desarrollo de algoritmos y técnicas que permiten a las computadoras entender, interpretar y generar lenguaje natural, así como interactuar con los seres humanos de una manera natural."
# Cargar el modelo de idioma de spaCy
nlp = spacy.load('es_core_news_sm', disable=['parser', 'ner'])
# Crear un objeto Doc para el texto
doc = nlp(texto)
# Tokenización de palabras
palabras = [token.text for token in doc]
cont=0
print("Palabras:")
for palabra in palabras:
    print(palabra)
    cont=cont+1
end_time = time.time()
print("\n Tiempo:", end_time - start_time, "seconds")
esperado=63
efect=((cont/esperado)*100)
print("palabras encontradas:",cont)
print("palabras esperadas:", esperado)
print("Efectivdad de tokenización", efect)
```

```python
import time
import stanfordnlp
start_time = time.time()
texto = "NLP, o Procesamiento del Lenguaje Natural, es una subdisciplina de la inteligencia artificial que se enfoca en la interacción entre las computadoras y el lenguaje humano. Se trata del desarrollo de algoritmos y técnicas que permiten a las computadoras entender, interpretar y generar lenguaje natural, así como interactuar con los seres humanos de una manera natural."

# Descargar el modelo de idioma español
#stanfordnlp.download('es')

# Cargar el modelo de idioma español
nlp = stanfordnlp.Pipeline(lang='es', processors='tokenize')

# Procesar el texto
doc = nlp(texto)

# Tokenización de palabras
palabras = []
for sent in doc.sentences:
    for word in sent.words:
        palabras.append(word.text)

cont=0
print("Palabras:")
for palabra in palabras:
    print(palabra)
    cont=cont+1
end_time = time.time()
print("\n Tiempo:", end_time - start_time, "seconds")
esperado=63
efect=((cont/esperado)*100)
print("palabras encontradas:",cont)
print("palabras esperadas:", esperado)
print("Efectivdad de tokenización", efect)
```

```python
from allennlp.data.tokenizers import WhitespaceTokenizer
import time
start_time = time.time()
tokenizer = WhitespaceTokenizer()
text = "NLP, o Procesamiento del Lenguaje Natural, es una subdisciplina de la inteligencia artificial que se enfoca en la interacción entre las computadoras y el lenguaje humano. Se trata del desarrollo de algoritmos y técnicas que permiten a las computadoras entender, interpretar y generar lenguaje natural, así como interactuar con los seres humanos de una manera natural."
tokens = tokenizer.tokenize(text)
print(tokens)
end_time = time.time()
print("\n Tiempo:", end_time - start_time, "seconds")
print("Encontradas:",len(tokens))
esperado=63
efect=((len(tokens)/esperado)*100)
print("palabras esperadas:", esperado)
print("Efectivdad de tokenización", efect)
```
